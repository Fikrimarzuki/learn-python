```python
import numpy as np

x = np.linspace(-10, 10, 100)
cos = np.cos(x)
cos3 = np.cos(x)*3
sin = np.sin(x-90)
sin2 = np.sin(x)*2
```

```python
# code here
import matplotlib.pyplot as plot
fig, ax = plt.subplots(figsize=(8, 4))

ax.plot(x, cos, linestyle='solid', color='red', label='Cos x')
ax.plot(x, cos3, linestyle="dashed", color='green', label='3 Cos x')
ax.plot(x, sin, linestyle='dashdot', color='blue', label='Sin (x-90)')
ax.plot(x, sin2, linestyle='', marker='x', color='yellow', label='2 Sin x')
ax.set_xlabel('X-Axis Label')
ax.set_ylabel('Y-Axis Label')
ax.set_title('Grafik Cos Sin')
ax.legend(loc='center')
plt.show()
```
