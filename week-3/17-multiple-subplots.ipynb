```python
import matplotlib.pyplot as plt
import numpy as np
```

<h2>Soal 1: Multiple Subplots</h2>

Dengan data di bawah ini buatlah visualisasi seperti expected output:

```python
x = np.linspace(2*-np.pi, 2*np.pi, 200)
tan = np.tan(x)/10
cos = np.cos(x)
sin = np.sin(x)
```

```python
# code here
fig, ax = plt.subplots(ncols=3, nrows=1, sharex=True, sharey=True, figsize=(18, 10))
ax[0].plot(x, tan, label="tan x")
ax[1].plot(x, sin, color="red", label="sin x")
ax[2].plot(x, cos, color="green", label="cos x")
ax[0].set_title("Grafik tan x")
ax[1].set_title("Grafik sin x")
ax[2].set_title("Grafik cos x")
fig.legend(loc="upper center")
plt.show()
```

<h2>Soal 2: Nested Axis</h2>

Dengan data di bawah ini, buatlah visualisasi seperti expected output:

```python
x = np.linspace(2*-np.pi, 2*np.pi, 100)
y = np.cos(x)
y2 = np.cos(x**2)
y3 = np.cos(x**3)
y4 = np.cos(x**4)
y5 = np.cos(x**5)
```

```python
# code here
fig, ax = plt.subplots(figsize=(18, 10))
ax.plot(x, y, color="b")
ax.set_title("Grafik Cos X")
ins1 = ax.inset_axes([-0.1, -0.1, 0.27, 0.27])
ins1.plot(x, y, color="b")
ins1.plot(x, y2, color="r")
ins1.set_title("Grafik Cox X^2")
ins2 = ax.inset_axes([0.9, -0.1, 0.27, 0.27])
ins2.plot(x, y, color="b")
ins2.plot(x, y3, color="g")
ins2.set_title("Grafik Cox X^3")
ins3 = ax.inset_axes([0.9, 0.9, 0.27, 0.27])
ins3.plot(x, y, color="b")
ins3.plot(x, y4, color="black")
ins3.set_title("Grafik Cox X^4")
ins4 = ax.inset_axes([-0.1, 0.9, 0.27, 0.27])
ins4.plot(x, y, color="b")
ins4.plot(x, y5, color="y")
ins4.set_title("Grafik Cox X^5")
plt.show()
```
